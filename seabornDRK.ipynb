```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Exercise_Data.csv")

print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pulse_cols = ["1 min", "15 min", "30 min"]

plt.figure()
sns.heatmap(df[pulse_cols].corr(), annot=True)

plt.title("Correlation Heatmap of Pulse Measurements")
plt.xlabel("Pulse Time Intervals")
plt.ylabel("Pulse Time Intervals")

plt.show()
```

```python
plt.figure()

sns.boxplot(data=df, x="diet", y="1 min", hue="kind")

plt.title("Pulse Rate (1 min) by Diet and Exercise Type")
plt.xlabel("Diet Type")
plt.ylabel("Pulse Rate (1 min)")

plt.legend(title="Exercise Type")

plt.show()
```

```python
df_melted = df.melt(id_vars=["diet", "kind"], 
                    value_vars=["1 min", "15 min", "30 min"],
                    var_name="Time",
                    value_name="Pulse")

plt.figure()

sns.boxplot(data=df_melted, x="Time", y="Pulse", hue="diet")

plt.title("Pulse Rate Over Time by Diet")
plt.xlabel("Time Interval")
plt.ylabel("Pulse Rate")

plt.legend(title="Diet Type")

plt.show()
```

```python
planets = sns.load_dataset("planets")
```

```python
# scatter
plt.figure()
sns.scatterplot(data=planets, x="distance", y="orbital_period")
plt.title("Distance vs Orbital Period")
plt.xlabel("Distance")
plt.ylabel("Orbital Period")
plt.show()

# line
plt.figure()
sns.lineplot(data=planets, x="year", y="mass")
plt.title("Mass Over Time")
plt.xlabel("Year")
plt.ylabel("Mass")
plt.show()
```

```python
# histogram
plt.figure()
sns.histplot(planets["mass"], kde=True)
plt.title("Distribution of Planet Mass")
plt.xlabel("Mass")
plt.ylabel("Count")
plt.show()

# KDE
plt.figure()
sns.kdeplot(planets["distance"], fill=True)
plt.title("Density of Planet Distance")
plt.xlabel("Distance")
plt.ylabel("Density")
plt.show()
```

```python
# boxplot
plt.figure()
sns.boxplot(data=planets, x="method", y="orbital_period")
plt.xticks(rotation=45)
plt.title("Orbital Period by Discovery Method")
plt.xlabel("Method")
plt.ylabel("Orbital Period")
plt.show()

# countplot
plt.figure()
sns.countplot(data=planets, x="method")
plt.xticks(rotation=45)
plt.title("Count of Planets by Discovery Method")
plt.xlabel("Method")
plt.ylabel("Count")
plt.show()
```
